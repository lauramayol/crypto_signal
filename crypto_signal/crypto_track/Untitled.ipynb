```python
from pytrends.request import TrendReq
import pandas
import datetime
```

```python
kw_list = ['buy bitcoin','btc usd']

pytrends = TrendReq(hl='en - US')
```

```python
pytrends.build_payload(kw_list, cat=0, timeframe='2017-01-18 2017-5-18', geo='', gprop='')
```

```python
df = pytrends.interest_over_time()
```
